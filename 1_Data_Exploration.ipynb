## Draw the visualizations with seaborn package

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print("Some Statistics of the Housing Price:\n")
print(df['SalePrice'].describe())
print("\nThe median of the Housing Price is: ", df['SalePrice'].median(axis = 0))
```

```python
sns.set(style="white")
sns.distplot(df['SalePrice'], kde = False, color = sns.color_palette("BuGn_r")[1], hist_kws={'alpha': 0.9})
plt.title("Overall Distribution of Sale Price")
plt.show()
```

```python
sns.set(style="white")
sns.distplot(np.log(df['SalePrice']), kde = False, color = sns.color_palette("Blues")[4], hist_kws={'alpha': 0.9})
plt.title("Sale price overall distribution after log transformation")
plt.show()
```

<font color = "blue">

## Time Series Exploration with SalePrice

```python
sns.factorplot('Yr Sold', 'SalePrice', estimator = np.mean, 
               size = 4.5, aspect = 1.4, data = df, color = "darkgreen")
plt.title("Effect of Year on SalePrice")
plt.show()
```

```python
sns.factorplot('Mo Sold', 'SalePrice', estimator = np.mean, 
               size = 4.5, aspect = 1.4, data = df,color = "darkblue")
plt.title("Effect of Month on SalePrice")
plt.show()
```

<font color = "blue">

## Numerical Data Exploration

```python
corr = df.select_dtypes(include = ['float64', 'int64']).iloc[:, 1:].corr()
plt.figure(figsize=(12, 12))
sns.heatmap(corr, vmax=1, square=True)
plt.title("Heatmap of the Numerical Predictors and SalePrice")
```

```python
cor_dict = corr['SalePrice'].to_dict()
del cor_dict['SalePrice']
print("List the numerical features decendingly by their correlation with Sale Price:\n")
for ele in sorted(cor_dict.items(), key = lambda x: -abs(x[1])):
    print("{0}: \t{1}".format(*ele))
```

```python
#plot the scatter plot of the SalePrice and the four most relevant numerical predictors 
plt.figure(1)
f, axarr = plt.subplots(2, 2, figsize=(10, 9))
price = df.SalePrice.values
axarr[0, 0].scatter(df["Overall Qual"].values, price, color = sns.color_palette("Blues")[2])
axarr[0, 0].set_title('Overall Qual')
axarr[0, 1].scatter(df["Gr Liv Area"].values, price, color = sns.color_palette("Blues")[4])
axarr[0, 1].set_title('Gr Liv Area')
axarr[1, 0].scatter(df["Garage Cars"].values, price, color = sns.color_palette("Blues")[3])
axarr[1, 0].set_title('Garage Cars')
axarr[1, 1].scatter(df['Garage Area'].values, price, color = sns.color_palette("Blues")[5])
axarr[1, 1].set_title('Garage Area')
f.text(-0.01, 0.5, 'Sale Price', va='center', rotation='vertical', fontsize = 12)
plt.tight_layout()
plt.show()
```

```python
#plot the relationship between SalePrice and YearBuilt, Year Remod/Add
fig = plt.figure(2, figsize=(9, 7))
plt.subplot(211)
plt.scatter(df["Year Built"].values, price)
plt.title('YearBuilt')

plt.subplot(212)
plt.scatter(df["Year Remod/Add"].values, price)
plt.title('YearRemodAdd')

fig.text(-0.01, 0.5, 'Sale Price', va = 'center', rotation = 'vertical', fontsize = 12)

plt.tight_layout()
```

## Categorical Data Exploration

```python
print(df.select_dtypes(include=['object']).columns.values)
```

```python
## group all neighborhoods with fewer than 50 records into one category
## and name it "other"
cnts = pd.DataFrame(df.groupby('Neighborhood').Order.count()).reset_index()
cnts.columns=['Neighborhood', 'cnts']
df = df.merge(cnts, on = 'Neighborhood')
df['Neighborhood_new'] = df.Neighborhood
df.ix[df.cnts < 150, 'Neighborhood_new'] = 'Other'
```

```python
df
```

```python
#plot the histogram of SalePrice and neighorhoods 
plt.figure(figsize = (12, 6))
sns.countplot(x = 'Neighborhood', data = df)
xt = plt.xticks(rotation=45)
```

```python
#plot the boxplot of SalePrice and the new neighorhoods after grouping 
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'Neighborhood_new', y = 'SalePrice',  data = df)
xt = plt.xticks(rotation=45)
plt.title("Effect of Neighborhood on SalePrice")
```

```python
#plot the violinplot of SalePrice and Sale Type
fig, ax = plt.subplots(1, 1, figsize = (10, 3))
plt.title("Effect of Sale Type on SalePrice")
sns.violinplot(x = 'Sale Type', y = 'SalePrice', data = df, ax = ax)
plt.show()
```

```python
#plot the violinplot of SalePrice and Sale Condition
fig, ax = plt.subplots(1, 1, figsize = (10, 3))
plt.title("Effect of Sale Condition on SalePrice")
sns.violinplot(x = 'Sale Condition', y = 'SalePrice', data = df, ax = ax)
plt.show()
```

```python
#plot the violinplot of SalePrice and time (including month sold and year sold)
g = sns.FacetGrid(df, col = 'Yr Sold', col_wrap = 3)
g.map(sns.boxplot, 'Mo Sold', 'SalePrice', palette='Set2', order = range(1, 13)).set(ylim = (0, 500000))
plt.tight_layout()
```
